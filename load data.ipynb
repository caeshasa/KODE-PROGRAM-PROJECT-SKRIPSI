```python
import pandas as pd

fact = pd.read_csv("fact_tables.csv", sep=",", engine="python")
print(fact.shape)
print(fact.columns.tolist())
fact.head(2)
```

```python
import re

fact.columns = [
    re.sub(r"[^\w]+", "_", c).strip("_")[:128]
    for c in fact.columns
]

fact.to_sql("fact_tables", con=engine, if_exists="replace", index=False)
```
